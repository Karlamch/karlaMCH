```python
#1.Establezca una variable dependiente ( Y ) y una variable independiente ( X )

#Variable independiente: examenes cortos
#Variable dependiente: examen final
import numpy as np
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Karlamch/karlaMCH/refs/heads/main/Salary_dataset.csv')
df.head()
X = df['YearsExperience']
Y = df['Salary']


#2.Realice un diagrama de dispersión para estos datos
import matplotlib.pyplot as plt #ayuda a graficar
plt.scatter(X,Y, color="purple") #son los puntos
plt.xlabel('Examenes cortos') #indica el nombre de Y y X
plt.ylabel('Examen final')
plt.title('Grafico de dispersión')
plt.grid()

#3.¿Los datos soportan la suposición de linealidad?
# Si se puede trazar una linea

#4, Calcule el coeficiente de correlación e interprete el resultado.
Sxx = np.sum((X-np.mean(X))**2)
Sxy = np.sum((X-np.mean(X))*(Y-np.mean(Y)))
Syy = np.sum((Y-np.mean(Y))**2)
r = Sxy/np.sqrt(Sxx*Syy)
print("Coeficiente de correlación: ", r)

#5. Calcule el coeficiente de determinación e interprete el resultado.
r2 = r**2
print("Coeficiente de determinación: ", r2)
# que el 74.75% justifica la variabilidad de X y Y

#6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
b1 = Sxy/Sxx
b0 = np.mean(Y) - b1*np.mean(X)
print(f"Pendiente: {b1}")
print(f"Intercepto: {b0}")

Yc = b1 * X + b0

import matplotlib.pyplot as plt
plt.scatter(X,Y, color="purple")
plt.plot(X,Yc, color="red")
plt.xlabel('Examenes cortos')
plt.ylabel('Examen final')
plt.title('Grafico de dispersión')
plt.grid()

#7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de
#regresión ajustada ( b1 )
nivel_de_significancia = 0.05
import scipy.stats as st
t = st.t.ppf(1-nivel_de_significancia/2, len(X)-2)
se_b1 = np.sqrt(np.sum((Y-Yc)**2)/(len(Y)-2)) / np.sqrt(Sxx)
confianza_b1 = (b1-t*se_b1, b1+t*se_b1)
print(f"Intervalo de confianza para b1: {confianza_b1}")

#8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
#¿Parece que se verifican los supuestos?
residuales = Y - Yc
plt.figure()
plt.scatter(X,residuales, color="blue")
plt.xlabel("Examenes cortos")
plt.ylabel("Residuales")
plt.title("Grafico de residuales")
plt.axhline(y=0, color = "red", linestyle = "--")

#9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f"valor_p de shapiro : {valor_p_sh}")
#No se descarta la hipotesis nula, los numeros si van bien

#10. Realice la prueba de Brausch-Pagan para los residuales y comente el resultado.

#11. Tres estudiantes sacaron 70, 75 y 84 de calificación. Según la recta de regresión ajustada, ¿cuáles son los resultados esperados para estos tres alumnos?
print(f"Yc para 70: {b1*70+b0}")
print(f"Yc para 75: {b1*75+b0}")
print(f"Yc para 84: {b1*84+b0}")

#12. Realice una tabla ANOVA e interprete el resultado.
print(f"Sxx: {Sxx}")
print(f"Sxy: {Sxy}")
print(f"Syy: {Syy}")
print(f"SSE: {np.sum(residuales**2)}")
print(f"SST: {np.sum((Y-np.mean(Y))**2)}")
print(f"SSR: {np.sum((Yc-np.mean(Y))**2)}")
print(f"MSR: {np.sum((Yc-np.mean(Y))**2)/1}")
print(f"MSE: {np.sum(residuales**2)/(len(Y)-2)}")
print(f"F: {(np.sum((Yc-np.mean(Y))**2)/1)/(np.sum(residuales**2)/(len(Y)-2))}")
print(f"p-value: {1-st.f.cdf((np.sum((Yc-np.mean(Y))**2)/1)/(np.sum(residuales**2)/(len(Y)-2)),1,len(Y)-2)}")
```

#**"Salary Dataset simple linear regression."**

##Introducción
El análisis de datos es una herramienta fundamental en la ciencia de datos para extraer información significativa y realizar predicciones sobre situaciones futuras. En este contexto, la regresión lineal simple se utiliza para modelar la relación entre dos variables, una independiente y una dependiente, con el objetivo de predecir el valor de la variable dependiente. El "Salary Dataset" es un conjunto de datos que contiene información sobre los salarios de los empleados en función de su experiencia laboral, lo que lo convierte en un excelente caso de estudio para aplicar regresión lineal simple. El objetivo de este análisis es estudiar cómo la experiencia laboral (en años) afecta el salario de los individuos. Este análisis puede ser útil para empresas y analistas de recursos humanos para predecir salarios futuros y tomar decisiones sobre remuneraciones.

##Marco teórico:
La regresión lineal simple es una técnica estadística que establece una relación lineal entre dos variables. En este caso, se busca modelar la relación entre el salario de una persona (variable dependiente) y su experiencia laboral (variable independiente). El modelo de regresión lineal simple se expresa de la siguiente forma:


El modelo de regresión lineal simple se expresa como:

Y= B0 + B1X + E

Donde:
Y = es la variable dependiente (salario).

X = es la variable independiente (experiencia laboral en años).

B0 = es la intersección con el eje Y (El valor de Y cuando X= 0)

B1 = es la pendiente de la línea de regresión, que indica cuánto cambia el salario por cada año adicional de experiencia.

E = es el término de error, que refleja la variabilidad no explicada por el modelo

El modelo tiene como objetivo encontrar los parámetros B0 y B1 que minimicen el error cuadrático medio, lo que nos permite hacer predicciones precisas.

##Hipótesis
La hipótesis que se plantea en este estudio es que la experiencia laboral tiene una relación positiva y significativa con el salario de los individuos. Es decir, a medida que aumenta la experiencia laboral, se espera que el salario también aumente. La regresión lineal simple puede ayudar a confirmar o refutar esta hipótesis al identificar la fuerza y la dirección de la relación entre ambas variables.


##Metodología
Los pasos metodológicos son los siguientes:

- Recopilación y limpieza de datos:
Se descargó el archivo CSV desde el enlace proporcionado. Posteriormente, se inspeccionaron y limpiaron los datos, asegurándose de que no hubiera valores nulos ni erróneos.

- Análisis exploratorio de datos (EDA):
Se realizó un análisis preliminar para visualizar la distribución de las variables y entender la relación entre la experiencia laboral (X) y el salario (Y). Se utilizaron herramientas estadísticas y gráficas, como diagramas de dispersión, para identificar patrones en los datos.

- Ajuste del modelo de regresión lineal:
Se aplicó un modelo de regresión lineal simple utilizando la variable experiencia laboral (X) como predictor del salario (Y). Se estimaron los coeficientes del modelo utilizando el método de mínimos cuadrados.

- Evaluación del modelo:
Se evaluó el modelo utilizando métricas estadísticas como el coeficiente de determinación r2 el error cuadrático medio (MSE) y el análisis de residuos, con el fin de verificar la precisión y la validez del modelo.

Se realizaron predicciones sobre salarios futuros basados en la experiencia laboral y se analizaron los resultados obtenidos.
