```python
import pandas as pd
import numpy as np

df =pd.read_csv("https://raw.githubusercontent.com/Karlamch/karlaMCH/refs/heads/main/USA_cars_datasets.csv")
df.dropna(inplace = True)
df
```

# INTRODUCCIÓN


Se presenta un conjunto de datos de anuncios de autos usados ​​de EE. UU. Los datos incluyen información como precio, marca, modelo, año, estado del título y kilometraje. El propósito de recopilar estos datos podría ser para diversos análisis, como comprender las tendencias de precios de los autos, la distribución de las diferentes marcas y modelos en el mercado de autos usados ​​o la relación entre el kilometraje y el precio.

# HIPÓTESIS


Los vehículos más recientes y con menor kilometraje tienden a tener precios de venta más altos en el mercado de automóviles usados ​​de EE.UU. UU. Además, la marca del vehículo podría influir significativamente en el precio, incluso controlando por la antigüedad y el kilometraje.



# MARCO TEÓRICO


Se podría enmarcar dentro de la teoría económica de la depreciación de activos y la valoración de mercado .

* Depreciación: Los automóviles son bienes duraderos que generalmente
pierden valor con el tiempo y el uso (kilometraje). La tasa de depreciación puede variar según la marca, el modelo, la condición y otros factores.

* Valoración de Mercado: El precio de un automóvil usado está determinado por la interacción de la oferta y la demanda. Factores como la reputación de la marca (confianza en la calidad y durabilidad), la demanda del modelo específico, la condición del vehículo (indicada por el estado del título y el kilometraje), la antigüedad y las características del vehículo influyen en lo que los compradores están dispuestos a pagar.

* Teoría del Capital Humano (implícita): Los compradores pueden estar dispuestos a pagar más por vehículos más nuevos y con menos kilometraje para reducir los costos de mantenimiento y reparación futuros, esencialmente invirtiendo en un "capital humano" del vehículo más alto.
